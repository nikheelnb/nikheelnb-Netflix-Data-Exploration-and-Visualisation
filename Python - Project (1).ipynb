# Q. 1  - Defining Problem Statement and Analysing basic metrics

Defining Problem Statment:

    With over 200 million subscribers worldwide, Netflix is one of the largest streaming platforms. 
    In order to keep its subscribers engaged, Netflix continuously adds new movies and TV shows 
    to its platform. 
    However, with the abundance of options, it can be challenging for subscribers to find content 
    that they are interested in. 
    As a result, Netflix needs to understand what types of content its subscribers are interested 
    in and how to make that content more discoverable.

Basic Metrics for Netflix Data Exploration :

    1 - Subscriber Growth: One of the most important metrics for Netflix is its subscriber growth. 
    Netflix needs to continuously attract new subscribers in order to maintain its revenue and 
    increase its market share.
    By analyzing subscriber growth over time, Netflix can identify trends and make strategic,
    decisions to attract and retain subscribers.
    
    2 - Content Popularity: Another key metric for Netflix is the popularity of its content.
    By analyzing viewing data, Netflix can determine which movies and TV shows are most popular
    among its subscribers.This information can help Netflix make decisions about which content 
    to acquire and how to promote that content on its platform.

    3 - Engagement Metrics: Netflix can also analyze engagement metrics, 
    such as the number of hours watched per user, to understand how its subscribers are using its platform. 
    By understanding how subscribers engage with its platform, 
    Netflix can improve the user experience and create more personalized recommendations.
    
    4 - Retention Rate: Retaining subscribers is just as important as acquiring new ones. 
    By analyzing retention rates, Netflix can identify areas where it needs to improve its 
    platform or content in order to keep subscribers engaged and subscribed.
    
    5 - Revenue Metrics: Finally, revenue metrics, such as average revenue per user and lifetime 
    customer value, are important for Netflix's financial health.By analyzing these metrics,
    Netflix can make decisions about pricing and marketing strategies to maximize revenue.

By analyzing these basic metrics, Netflix can gain insights into its subscribers' behavior and 
preferences, and use that information to make strategic decisions about content acquisition, 
promotion, and user experience.

```python
# Top 10 Countries - Ratings of Movies

expanded_country_list = data['country'].apply(lambda x : str(x).split(',')).tolist()
netflix_country_list = pd.DataFrame(expanded_country_list, index = data['title'])
netflix_country_list = netflix_country_list.stack()
netflix_country_list = pd.DataFrame(netflix_country_list.reset_index())
netflix_country_list.drop(['level_1'], axis = 1, inplace = True)
netflix_country_list.rename(columns = {0:'Country'}, inplace = True)
netflix_country_expanded = netflix_country_list.merge(data[['show_id', 'type', 'title', 
                         'director', 'country', 'date_added','release_year', 'rating', 
                         'duration', 'listed_in', 'description']], on = ['title'], how = 'left')
netflix_country_expanded['Country'] = netflix_country_expanded['Country'].str.strip()
netflix_country_count = netflix_country_expanded['Country'].value_counts()
```

```python
netflix_country_top10 = netflix_country_expanded[netflix_country_expanded['Country']
                        .isin(list(netflix_country_count.index)[:11])]
netflix_country_top10 = netflix_country_top10[['Country','rating']]
netflix_country_top10['Count'] = 1
```

```python
df=netflix_country_top10.pivot_table(index="Country",
                     columns="rating",
                     aggfunc=["count"],
                     fill_value=0)
column_names = []
for x in df.columns:
  column_names.append(x[2])
df.columns = column_names
df = df.loc[['United States',
 'India',
 'Not Available',
 'United Kingdom',
 'Canada',
 'France',
 'Japan',
 'Spain',
 'South Korea',
 'Germany',
 'Mexico']]
```
